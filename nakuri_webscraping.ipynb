```python
# importing the necessary libraries
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
```

```python
def web_driver():
  options = webdriver.ChromeOptions()
  options.add_argument("--verbose")
  options.add_argument("--headless")
  options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.0.0 Safari/537.36")
  options.add_argument("--no-sandbox")
  options.add_argument("--disable-gpu")
  options.add_argument("--window-size = 1920, 1200")

  driver = webdriver.Chrome(options = options)
  return driver
```

```python
driver = web_driver()
```

```python
# webscraping through the nakuri
url = "https://www.naukri.com/data-science-jobs"
driver.get(url)
driver.implicitly_wait(10)
```

```python
html_content = driver.page_source

soup = BeautifulSoup(html_content, "html.parser")

# extracting the job title
team_elements =soup.find_all("a", class_="title")
job_titles = []
for job_listing in team_elements:
  job_title = job_listing.text.strip()
  job_titles.append(job_title)
for job_title in job_titles:
  print(job_title)
```

```python
# extracting the experenice of the data science job and storing in the list
team_elements1 =soup.find_all("span", class_="expwdth")
experience = []
for job_listing1 in team_elements1:
    experince = job_listing1.text.strip()
    experience.append(experince)
for experinces in experience:
  print(experinces)
```

```python
#extracting the location of the datascience

team_elements2 = soup.find_all("span", class_="locWdth")
locations = []
for job_listing2 in team_elements2:
  location = job_listing2.text.strip()
  locations.append(location)
for location in locations:
  print(location)
```

```python
#Extracting the company name of the datascience job
team_elements3 = soup.find_all("a",class_="comp-name mw-25")
companies = []
for job_listing3 in team_elements3:
  company = job_listing3.text.strip()
  companies.append(company)
for company in companies:
  print(company)
```

```python
#etrxating the slaries of the data science jobs

team_elements4 = soup.find_all("span",class_="sal-wrap ver-line")
salaries = []
for job_listing4 in team_elements4:
  salery = job_listing4.text.strip()
  salaries.append(salery)
for salary in salaries:
  print(salary)
```

```python
import numpy as np
max_length = max(len(job_titles),len(experience),len(locations),len(companies),len(salaries))
 # extend the shorter lists withe place holder values (NaN)
job_titles += [np.nan] * (max_length - len(job_titles))
experience += [np.nan] * (max_length - len(experience))
locations += [np.nan] * (max_length - len(locations))
companies += [np.nan] * (max_length - len(companies))
salaries += [np.nan] * (max_length - len(salaries))
```

```python
   # Append the scraped data to the list
data={
        "job_title": job_titles,
        "experience": experience,
        "location": locations,
        "company" : companies,
        "salary" : salaries
    }
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt


#  Get the salary data
location_data = df["location"]

# Count the occurrences of each salary value
location_counts = location_data.value_counts()

# Create a bar chart of the salary data

plt.xticks(rotation=90)
plt.bar(location_counts.index, location_counts.values, edgecolor="black")

plt.xlabel("location",fontsize=28)
plt.ylabel("number of jobs")
plt.title("Distribution of locations for Data Science Jobs on Naukri")
plt.show()
```
